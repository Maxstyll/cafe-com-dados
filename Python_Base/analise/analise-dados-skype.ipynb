## Dicionario do dataset
* **SK_ID_CURR**: ID do empréstimo em nossa amostra
* **TARGET**: Variável alvo (1 - cliente com dificuldades de pagamento: ele/ela atrasou mais de X dias no pagamento em pelo menos uma das primeiras Y parcelas do empréstimo em nossa amostra, 0 - todos os outros casos)

## Retirar registros que tenha variaveis com outliers (Escor Z)

```python
des_target = df["TARGET"].unique()

name_columns = []
name_columns.extend(integer_features)
name_columns.extend(float_features)
```

```python
%%time
for variavel in name_columns:
    df_defaut = []
    for value, target in enumerate(des_target):
        dados = df[df["TARGET"]==value]
        print(f"Classe: {des_target[value]} - variavel: {variavel}")

        data_mean, data_std = dados[variavel].mean(), dados[variavel].std()
        data_min, data_max = dados[variavel].min(), dados[variavel].max()
        print("Real Min: %.3f Real Max: %.3f" %  (data_min, data_max))

        cut_off = data_std * 3
        lower, upper = data_mean - cut_off, data_mean + cut_off
        print("Limit Min: %.3f Limit Max: %.3f" %  (lower, upper))


        dataset = dados[dados[variavel] >= lower]
        if len(dataset) == 0:
            dataset = dados[dados[variavel] <= upper]
        else:
            dataset = dataset[dataset[variavel] <= upper]


        df_defaut.append(dataset)

        outliers = pd.concat([dados[dados[variavel] < lower], dados[dados[variavel] > upper]])
        print("Identfied outliers: %d \n" % len(outliers))

    df = pd.concat(df_defaut)
```

## Normalizar as variaveis

```python
numeric_transformer = []

def norm_variaveis (colunm):
    df_smote["NORM_" + colunm] = df_smote[colunm] / df_smote[colunm].abs().max()
    numeric_transformer.append("NORM_" + colunm)
    
```

```python
for variavel in float_features:
    norm_variaveis(variavel)
```

## Identificar as melhores relações de variaveis para a variavel taget (random forest)

```python
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
name_columns_relevantes = []
name_columns_relevantes.extend(name_columns)
name_columns_relevantes.extend(name_category_features)

df_relevantes = df_scaled[name_columns_relevantes].copy()
```

```python
df_relevantes.head()
```

```python
%%time
name_columns = numeric_transformer
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant')),
    ('one_hot', OneHotEncoder())])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, name_columns),
        ('cat', categorical_transformer, name_category_features)
    ])
```

```python
name_columns
```

```python
name_category_features
```

```python
%%time
pipe = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier',  RandomForestClassifier(random_state=5, n_estimators=10))])
    
model = pipe.fit(df_relevantes, y)
```

```python
onehot_columns = list(pipe.named_steps['preprocessor'].named_transformers_['cat'].named_steps['one_hot'].get_feature_names(input_features=name_category_features))
numeric_features_list = list(name_columns)
numeric_features_list.extend(onehot_columns)
```

```python
import eli5
```

```python
eli5.explain_weights(pipe.named_steps['classifier'], top=22, feature_names=numeric_features_list)
```
